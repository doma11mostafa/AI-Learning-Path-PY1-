# The while Loop

The loop has two parts: 
(1) a condition that is tested for a true or false value, and
(2) a block (statement or set of statements) that is repeated as long 


```python
#Ex
x=1
while x<4:
    print(x**2)
    x+=1 # increment x by +1 to didn't be in 
```

```python
#Difference  Between loops
for i in range (10):
    print(i)
print('***********************')
x=0
while x<10:
    print(x)
    x+=1
```

# The while Loop Is a Pretest Loop


```python
ans='Y'
# loop will repeat itself as long as  ans="y" 
while ans=="Y":
    sal=eval(input("Enter the salary: " ))
    tax=0.05*sal
    print(tax)
    ans=input("If u wanna continue please iner Y OR N: ")
print('Thanks')    
```

# Sentinels

```python
sal=eval(input("Enter the salary, -1 to stop: "))
while sal !=-1:
    tax=0.05*sal
    print("The tax will be:",tax )
    sal=eval(input('Enter the salary, -1 to stop:'))
print("Thank you!")
```

# The break Statement

```python
x = 1
while x < 5 :
    print (x)
    x = x + 1
    break
else :
    print ('got here because the condition is not satisfied! ', x)
print("Done!")
```

# The continue Statement

```python
i=0
while i<6:
    i+=1
    if i==3:
        continue
    print(i)    
```

```python
i=0
while i<6:
    i+=1
    if i==3:
        break
    print(i) 
```

# Nested loops

```python
for i in range(1,11):
    for j in range(1,11):
        print(i*j, end=" ")
    print('')
```

# Nested while Loop


```python
i=1
while i<=10:
    j=1
    while j<=10:
          print (i*j, end=' ')
          j+=1
    i+=1
    print('')
```

```python
#first Convert into a list
my_tuple = ("apple", "banana", "cherry")
y = list(my_tuple)
y.append("orange")
my_tuple = tuple(y)
print(my_tuple)

```

# Dictionaries

In Python, a dictionary is an object that stores a collection of data. Each element that is stored in a dictionary has two parts: a key and a value. Key-value pairs are often referred to as mappings because each key is mapped to a value.


```python
phonebook = {'Jamil':'963−11112222', 'Fadia':'692−43219876', 'Mazen':'961−3665169'}
print(phonebook)
```

```python
print(phonebook['Jamil'])
```

```python
phonebook = {'Jamil':'963−11112222', 'Fadia':'692−43219876', 'Mazen':'961−3665169'}
del phonebook['Fadia']
print(phonebook)
```
