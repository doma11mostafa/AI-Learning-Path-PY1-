# أمر fminbound

```python
 
import numpy as np
import scipy as sc


def f(x):
    return x**2 + 10*np.sin(x)

a = sc.optimize.fminbound(f,# function
                          x1 = -10,# first bound 
                          x2 = 10,#second bound
                          xtol = 0.01, # Max Tolerance
                          full_output = True,# details 
                          disp = 1 ) #show only numbers

print(a)
# Value of x , Value of y , 0 for convergance
# , number of iterations
(-1.3061484200406244, -7.94582288025181, 0, 11)
```

# ●	صيغة مختلفة

●	"method": "BFGS": This is a key-value pair within the minimizer_kwargs dictionary. It specifies the optimization method to be used as "BFGS." BFGS stands for Broyden-Fletcher-Goldfarb-Shanno, and it's a well-known optimization algorithm for finding the minimum of a scalar function of several variables.
●	The BFGS algorithm is an iterative method that approximates the inverse Hessian matrix (a matrix of second derivatives) to guide the search for the minimum of the objective function. It's a quasi-Newton method and is widely used for unconstrained optimization problems.


```python
import numpy as np
import scipy as sc


func = lambda x: np.cos(14.5 * x - 0.3) + (x + 0.2) * x
x0=[1.]

ret = sc.optimize.basinhopping(func, #function
                               x0 , # a value to start
                                # method = BFGS
                               minimizer_kwargs={"method": "BFGS"},
                               niter=200) #max no. if ierations
print("global minimum: x = %.4f, f(x0) = %.4f" % (ret.x, ret.fun))
```

# ●	نفس الأمر لمعادلة من بعدين

●	minimizer_kwargs is intended to store settings or arguments for a minimization algorithm.
●	
●	"method": "L-BFGS-B": This is one of the key-value pairs within the minimizer_kwargs dictionary. It specifies the optimization method to be used. In this case, the method being used is "L-BFGS-B." "L-BFGS-B" stands for Limited-memory Broyden-Fletcher-Goldfarb-Shanno with Bound constraints, which is a popular optimization algorithm used to find the minimum of a scalar function of several variables, subject to bound constraints on those variables
●	"jac": True: This is another key-value pair within the minimizer_kwargs dictionary. It specifies whether to compute and provide the Jacobian matrix (the gradient of the objective function) during the optimization process. Setting it to True means that the Jacobian will be computed and used in the optimization. The Jacobian helps the optimization algorithm make more informed decisions about how to adjust the parameters to reach the minimum.  


```python
import numpy as np
import scipy as sc


def func2d(x):
    f = np.cos(14.5*x[0]-0.3)+(x[1]+0.2)*x[1]+(x[0]+0.2)*x[0]
    df = np.zeros(2)
    return f, df

x0 = [1.0, 1.0]
ret = sc.optimize.basinhopping(
        func2d,
        x0, 
        minimizer_kwargs={"method":"L-BFGS-B", "jac":True},
        niter=2000)

print("global minimum: x = [%.4f, %.4f], f(x0) = %.4f" % 
      (ret.x[0],ret.x[1],ret.fun))
```

# ●	أمر curve_fit  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
#function
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
#random values of X 
xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
```

In Python, the random.seed() function is used to initialize the random number generator.
When you set a seed value using random.seed(), it ensures that the sequence of random numbers generated by the random number generator is reproducible. This means that if you set the same seed value, you will get the same sequence of random numbers every time you run your code. This can be useful for debugging, testing, or ensuring that your results are consistent across different runs of your program.

```python
np.random.seed(1729)
# y noise
y_noise = 0.2 * np.random.normal(size=xdata.size)
ydata = y + y_noise
plt.plot(xdata, ydata, 'b-', label='data')  # draw the blue line

popt, pcov = curve_fit(func, xdata, ydata) #fitting
plt.plot(xdata, func(xdata, *popt), 'r-', #draw the red line
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))

#fit with bounds
popt, pcov =curve_fit(func,xdata,ydata,bounds=(0,[3., 1., 0.5]))
plt.plot(xdata, func(xdata, *popt),  #draw the green line
         'g--',label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
